```python
soup_page:list[BeautifulSoup] = [] 
for i in os.listdir('../temp/cheap_cars/'):
    full_path = os.path.join('../temp/cheap_cars',i)
    print(i)
    with open(full_path,'r') as file:
        content = file.read()
    soup_page.append(BeautifulSoup(content))
```

```python
css_selectors = {
    'ad':'a.dIkIJy',
    'title':'h3.hDGITe',
    'additional_info':'span.daYHkI',
    'car_info': 'div.WbYSG',
    'type_of_seller':'span.kWavwo:nth-of-type(1)',
    'location':'span.kWavwo[aria-label]',
    'price':'span.czQcuf',
}
```

```python
soup_page[0].select_one('a.dIkIJy').get('href')
```

```python
res = []
for page in soup_page:
    all_ads = page.select(css_selectors['ad'])
    print(len(all_ads))
    for car in all_ads:
        car_data = {}
        car_data['link'] = car.get('href')
        for name,selector in css_selectors.items():
            if name == 'ad':
                continue
            else:
                try:
                    car_data[name] = car.select_one(selector).text
                except:
                    car_data[name] = 'N/A'
        res.append(car_data)
                
```

```python
def get_make(title):
    make =  ' '.join([i.replace('\n','') for i in title.split(' ')[0:2]])
    if '(reserviert)' in make:
        make = ' '.join([i.replace('\n','') for i in title.split(' ')[1:3]])
    if 'Alfa Romeo' in make:
        make =  ' '.join([i.replace('\n','') for i in title.split(' ')[0:3]])
    return make
    
```

```python
fuel_types = ['Diesel','Benzin','Elektro','Hybrid Elektro/Benzin','Hybrid Elektro/Diesel']
transmission_types = ['Automatik','Schaltgetriebe']
waranty_type = ['Gewährleistung','Garantie']
registered = ['gültiges Pickerl',]
```

```python
def process_aditional_info(additional_info):
    info = additional_info.split(',')
    d = {}
    for i in info:
        if i.strip() in fuel_types:
            d['Fuel']=i
        if i.strip() in transmission_types:
            d['Transmission'] = i
        if i.strip() in waranty_type:
            d['Warranty'] = True
        if i.strip() in registered:
            d['Registered'] = True
    return d
```

```python
df['additional_info_dict'] = df.additional_info.apply(process_aditional_info)
df.additional_info_dict
```

```python
def process_car_info(car_info):
    car_info_add = car_info.split(' ')
    d = {}
    try:
        d['year']= int(car_info_add[0])
    except:
        d['year']  = 'N/A'
    if str(car_info_add[1]).startswith('EZ'):
        mileage = str(car_info_add[1]).replace('EZ','').replace('.','')
        mileage = int(mileage)
        d['mileage'] = mileage
    else:
        d['mileage'] = 'N/A'
    
    try:
        if 'km' in car_info_add[2]:
            hp = car_info_add[2].replace('km','')
            d['hp'] = int(hp)
        else:
            d['hp'] = 'N/A'
    except:
        d['hp'] = 'N/A'
    
    
    return d
    
```

```python
df['car_info_dict'] = df.car_info.apply(process_car_info)
```

```python
def get_zip_and_location(location):
    location_info = location.split(' ')
    d = {}
    d['ZipCode'] = int(location_info[0])
    d['County'] = ' '.join(location_info[1:])
    return d
```

```python
df['location_dict'] = df.location.apply(get_zip_and_location)
```
